```python
import pandas as pd
df = pd.read_csv("student_performance.csv")
df.shape, df.head()
```

```python
df.info()
df.isnull().sum()
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Final_Marks'], bins=20)
plt.title("Histogram of Final Marks")
plt.xlabel("Final Marks")
plt.ylabel("Count")
plt.show()

plt.scatter(df['Attendance'], df['Final_Marks'], alpha=0.6)
plt.title("Attendance vs Final Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Final Marks")
plt.show()

# Boxplot by Stream
streams = df['Stream'].unique()
stream_data = [df[df['Stream']==s]['Final_Marks'] for s in streams]
plt.boxplot(stream_data, labels=streams)
plt.title("Final Marks by Stream")
plt.show()
```

```python
df['Pass'] = (df['Final_Marks'] >= 40).astype(int)
print("Pass rate:", df['Pass'].mean())
df.groupby('Stream')['Final_Marks'].median().sort_values()
```

```python
# df.corr()[['Final_Marks']].sort_values(by='Final_Marks', ascending=False)

df.corr(numeric_only=True)[['Final_Marks']].sort_values(by='Final_Marks', ascending=False)
```
